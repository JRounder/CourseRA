```python
# determine lower, upper and inner quartile
q1 = df['ConvertedComp'].quantile(0.25)
q3 = df['ConvertedComp'].quantile(0.75)
iqr = q3 - q1
#identify outliers
high = q3+1.5*iqr
low = q1-1.5*iqr
```

```python
#identify outliers
dfout = df.loc[df['ConvertedComp'] > high]
dfout.head()
```

```python
# create df with no outliers
dfno = df.loc[(df['ConvertedComp'] < high) & (df['ConvertedComp'] > low)]
dfno.head()
```
